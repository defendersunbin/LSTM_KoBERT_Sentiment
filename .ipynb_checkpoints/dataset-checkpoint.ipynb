```python
import datetime
import FinanceDataReader as fdr  
import matplotlib.pyplot as plt
import pandas as pd
import openpyxl
```

```python
from pykrx import stock
import time
import pandas as pd
```

```python
kospi100 = pd.read_excel('코스피100(8.14).xlsx')
kospi100
```

```python
start = datetime.datetime(2021, 1, 1)
end = datetime.datetime(2021, 6, 30)
```

```python
stocks = fdr.StockListing('KOSPI') # 코스피
stocks
```

```python
stocks=stocks[['Code','Name']]
stocks
```

```python
merged_df = pd.merge(kospi100, stocks, on='Name', how='left')
merged_df 
```

```python
merged_df[merged_df['Code'].isnull()]
```

```python
merged_df=merged_df.dropna(axis=0)
merged_df
```

```python
for code, name in zip(merged_df['Code'],merged_df['Name']):
    stock_data = fdr.DataReader(code, start='2015-01-01', end='2023-08-08')
    stock_data.reset_index(inplace=True)
    print(stock_data)
    company_data=stock.get_market_fundamental_by_date(fromdate="20150101", todate="20230808", ticker=code)
    company_data.reset_index(inplace=True)
    company_data.rename(columns = {'날짜':'Date'}, inplace=True)
    print(company_data)
    df=pd.merge(stock_data, company_data, on='Date', how='left')
    df=df.fillna(method='ffill')
    name= f"{code}_{name}.csv"
    df.to_csv(name)
```
